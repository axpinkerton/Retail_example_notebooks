```python
res_df = pd.merge(data,hiers,left_on='product_id',right_on='product_id',how='left')
res_df
```

```python
res_res = res_df.groupby(['store_id','hierarchy2_id','date'])[['sales']].agg('sum')
res_res.reset_index(inplace=True)
res_res['unique_id'] = res_res['store_id']+"_"+res_res['hierarchy2_id']
res_res
```

```python
df1 = res_res.groupby(['store_id'])[['sales']].agg(sum).reset_index()
best_store = df1[df1['sales']==max(df1['sales'])][['store_id']]
top_store = best_store['store_id'].values[0]
```

```python
top_store_sls = res_res[res_res['store_id']==top_store].copy()
total_daily = top_store_sls.groupby(['store_id','date'])[['sales']].agg(np.sum).reset_index()
total_daily.to_csv('/Users/APinkerton/Downloads/top_store_daily.csv')
```

```python
depts = top_store_sls[top_store_sls['sales']>0].copy()
dept_sales_counts = depts.groupby(['hierarchy2_id'])[['sales']].agg('count').reset_index()
intermittent_dept = dept_sales_counts[dept_sales_counts['sales']<100].sort_values(by='sales',ascending=False)['hierarchy2_id'].values[0]

ts_intermittent_input = top_store_sls[top_store_sls['hierarchy2_id']==intermittent_dept].copy()
ts_intermittent_input.to_csv('/Users/APinkerton/Downloads/ts_intermittent_input.csv')
```

```python
## data zip folder: 
## https://www.kaggle.com/datasets/jazidesigns/time-series-data-sales

# New dataset--going to do bayesian structural time series forcasts for each item total and store total by month in R -- Cleaning the data here first

itm_cats = pd.read_csv('/Users/APinkerton/Downloads/archive (22)/item_categories.csv')
items = pd.read_csv('/Users/APinkerton/Downloads/archive (22)/items.csv')
sales = pd.read_csv('/Users/APinkerton/Downloads/archive (22)/sales_train.csv')
shops = pd.read_csv('/Users/APinkerton/Downloads/archive (22)/shops.csv')
test = pd.read_csv('/Users/APinkerton/Downloads/archive (22)/test.csv')
```

```python
sales['shop_id'] = sales['shop_id'].astype('category')
sales['item_id'] = sales['item_id'].astype('category')
day = []
mo = []
year = []

for x in sales['date']:
    day.append((x[0:2]))
    mo.append(x[3:5])
    year.append(x[6::])
sales['year'] = year
sales['month'] = mo
sales['day'] = day
sales['dt'] = sales['year']+"-"+sales['month']+'-'+sales['day']
sales['dt'] = pd.to_datetime(sales['dt'])
sales
```

```python
# Attributing the Year and Month to group by later
sls_df = sales[['dt', 'shop_id', 'item_id', 'item_price','item_cnt_day']].copy()
sls_df.columns = ['date', 'shop', 'item', 'sales_rev','sales_units']
sls_df['yr_mo'] = (sls_df['date'].dt.year).astype(str)+'-'+ (sls_df['date'].dt.month).astype(str)
sls_df
```

```python
store_df = sls_df.groupby(['shop','yr_mo'])[['sales_rev','sales_units']].agg(sum).reset_index()
item_df = sls_df.groupby(['item','yr_mo'])[['sales_rev','sales_units']].agg(sum).reset_index()
```

```python
store_df['yr_mo'] = pd.to_datetime(store_df['yr_mo'])
store_df.sort_values(by=['shop','yr_mo'],ascending=True,inplace=True)
store_df
```

```python
## How many zeroes are in each store's monthly total--too little data will yield poor forecasts 
store_df2 = store_df.groupby(['shop'])[['sales_rev','sales_units']].agg(lambda x: x.eq(0).sum()).reset_index()

## filtering out stores with too little data--i'm only keeping stores with at least 24 months to get at  least 2 full periods
final_store = store_df2[store_df2['sales_rev']<11].copy()
store_list = list(final_store['shop'])
```

```python
filt_stores = store_df[store_df['shop'].isin(store_list)].copy()
filt_stores
```

```python
item_df['yr_mo'] = pd.to_datetime(item_df['yr_mo'])
item_df.sort_values(by=['item','yr_mo'],ascending=True,inplace=True)

## 34 total months possible in data
print(max(item_df['yr_mo']))
print(min(item_df['yr_mo']))
```

```python
## How many zeroes are in each item's monthly total--too little data will yield poor forecasts 

item_df2 = item_df.groupby(['item'])[['sales_rev','sales_units']].agg(lambda x: x.eq(0).sum()).reset_index()
item_list = list(item_df2[item_df2['sales_rev']<11]['item'])
filt_item = item_df[item_df['item'].isin(item_list)].copy()
```
